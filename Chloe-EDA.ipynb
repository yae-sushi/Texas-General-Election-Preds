```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('Data Collection/combined_df.csv')
```

```python
df.head()
```

```python
# voting percentage compared to voting age population by county, CWWP hue
```

```python
plt.figure(figsize = (12, 5))
subset = df.sort_values(by='2012_voted_perc', ascending=False)[:10]
plt.subplot(1, 2, 1)
sns.barplot(data=subset, x = 'county', y = '2012_voted_perc')
plt.xticks(rotation = 65, );
plt.subplot(1, 2, 2)
sns.barplot(data=subset, x = 'county', y = '2012_unemp')
plt.xticks(rotation = 65, );
```

```python
plt.figure(figsize = (12, 5))
subset = df.sort_values(by='2016_voted_perc', ascending=False)[:10]
plt.subplot(1, 2, 1)
sns.barplot(data=subset, x = 'county', y = '2016_voted_perc')
plt.xticks(rotation = 65)
plt.subplot(1, 2, 2)
sns.barplot(data=subset, x = 'county', y = '2016_unemp')
plt.xticks(rotation = 65);
```

```python
plt.figure(figsize = (12, 5))
subset = df.sort_values(by='2020_voted_perc', ascending=False)[:10]
plt.subplot(1, 2, 1)
sns.barplot(data=subset, x = 'county', y = '2020_voted_perc')
plt.xticks(rotation = 65)
plt.subplot(1, 2, 2)
sns.barplot(data=subset, x = 'county', y = '2020_unemp')
plt.xticks(rotation = 65);
```
